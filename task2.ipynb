## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
print (img)
print (img.clip(T1,T2))
plt.imshow(img.clip(T1,T2), 'gray')
plt.colorbar()

'''img.clip -> Tells img to return a copy of itself using intensity clipping.
Given two thresholds ð‘‡1 and ð‘‡2, each pixel (ð‘¥, ð‘¦) of the image with an intensity value ð‘”(x,y)
less than the threshold ð‘‡1 is assigned the value ð‘‡1 and each pixel with a value
greater than ð‘‡2 is assigned the value ð‘‡2. Pixels with a value between the two thresholds
ð‘‡1 and ð‘‡2 remain unchange'''
```
