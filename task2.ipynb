## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
#print (img)
print (img.clip(T1,T2))
plt.imshow(img.clip(T1,T2), 'gray')
plt.colorbar()

'''img.clip -> Tells img to return a copy of itself using intensity clipping.
Given two thresholds 𝑇1 and 𝑇2, each pixel (𝑥, 𝑦) of the image with an intensity value 𝑔(x,y)
less than the threshold 𝑇1 is assigned the value 𝑇1 and each pixel with a value
greater than 𝑇2 is assigned the value 𝑇2. Pixels with a value between the two thresholds
𝑇1 and 𝑇2 remain unchange'''
```
